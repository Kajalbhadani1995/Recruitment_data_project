```python
cursor.execute("""
CREATE TABLE IF NOT EXISTS recruitment_data (
    `Applicant ID` INT PRIMARY KEY,
    `Application Date` VARCHAR(30),
    `Application Month` VARCHAR(10),
    `Application Year` INT,
    `First Name` VARCHAR(30),
    `Last Name` VARCHAR(30),
    Age INT,
    Gender VARCHAR(15),
    `Date of Birth` DATE,
    `Phone Number` VARCHAR(50),
    Email VARCHAR(50),
    Address VARCHAR(100),
    City VARCHAR(50),
    State VARCHAR(50),
    `Zip Code` INT,
    Country VARCHAR(100),
    `Education Level` VARCHAR(50),
    `Years of Experience` INT,
    `Desired Salary` DECIMAL(10,2),
    `Job Title` VARCHAR(100),
    Status VARCHAR(50),
    `Full Name` VARCHAR(50)
);
""")

print("Table Created Successfully!")
```

```python
for index, row in df.iterrows():
    sql = """
    INSERT INTO recruitment_data (
        `Applicant ID`, `Application Date`, `Application Month`, `Application Year`,
        `First Name`, `Last Name`, Age, Gender, `Date of Birth`, `Phone Number`,
        Email, Address, City, State, `Zip Code`, Country, `Education Level`,
        `Years of Experience`, `Desired Salary`, `Job Title`, Status, `Full Name`
    ) VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)
    """
    cursor.execute(sql, tuple(row))
connection.commit()
print("Data inserted successfully!")
```
